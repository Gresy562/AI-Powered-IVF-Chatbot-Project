Step 2: Import Libraries

```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Step 3: Load Dataset

```python
data = pd.read_csv(r"c:\Users\soujanya\OneDrive\Desktop\IVF_Chatbot_Dataset_Expanded.csv")
data.head()
```

Step 4: Prepare Data

```python
questions = data["question"]
answers = data["answer"]
```

```python
vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)
```

Step 5: Create Chatbot Function

```python
def chatbot(user_query):
    user_vector = vectorizer.transform([user_query])
    similarity = cosine_similarity(user_vector, question_vectors)
    index = similarity.argmax()
    return answers.iloc[index]
```

Step 6: Test Chatbot

```python
chatbot("What is IVF success rate?")
```

Step 7: Make It Interactive

```python
while True:
    query = input("Ask IVF Question (type exit to stop): ")
    if query.lower() == "exit":
        break
    print("Bot:", chatbot(query))
```
